## 0

Load the msft.csv dataset containing Microsoft Corporation stock prices and the pandas library.

```python
#TODO
#Load the msft.csv dataset containing Microsoft Corporation stock prices and the pandas library.
import pandas as pd
stocks = pd.read_csv('../input/msft.csv', parse_dates=['Date'])
#Print the first 5 rows of the dataset to understand its structure.
stocks.head()
```

## 1.

Create a new column "Change" that shows the daily change in the stock price between the closing price and the opening price.

```python
#TODO
# Create a new column "Change" that shows the daily change in the stock price between the closing price and the opening price.
stocks['Change'] = stocks['Close'] - stocks['Open']
# Display the first 5 rows of the updated DataFrame.
stocks.head()
```

## 2.

Create a new column "ChangeRel" that shows the daily relative change in the stock price with respect to the opening price.

```python
#TODO
#Create a new column "ChangeRel" that shows the daily relative change in the stock price with respect to the opening price.
stocks['ChangeRel'] = (stocks['Close'] - stocks['Open']) / stocks['Open'] * 100
#Display the first 5 rows of the updated DataFrame.
stocks.head()
```

## 3.

What is the largest percentage daily increase and decrease in price?

```python
#TODO
#What is the largest percentage daily increase and decrease in price?
max_increase = stocks['ChangeRel'].max()
max_decrease = stocks['ChangeRel'].min()
print(f"Largest percentage daily increase: {max_increase:.2f}%")
print(f"Largest percentage daily decrease: {max_decrease:.2f}%")
```

## 4.

Calculate the number of days with a price increase, the number of days where the price remained identical to the opening price, and the number of days with a price decrease?

```python
#TODO
#Calculate the number of days with a price increase, the number of days where the price remained identical to the opening price, and the number of days with a price decrease?
num_increase_days = (stocks['ChangeRel'] > 0).sum()
num_identical_days = (stocks['ChangeRel'] == 0).sum()
num_decrease_days = (stocks['ChangeRel'] < 0).sum()
print(f"Number of days with a price increase: {num_increase_days}")
print(f"Number of days with a price identical to the opening price: {num_identical_days}")
print(f"Number of days with a price decrease: {num_decrease_days}")
```

## 5.

Write a program that calculates the number of records in the dataset that relate to 2015 and those that relate to 2014.

```python
#TODO
#Write a program that calculates the number of records in the dataset that relate to 2015 and those that relate to 2014.
num_records_2015 = (stocks['Date'].dt.year == 2015).sum()
num_records_2014 = (stocks['Date'].dt.year == 2014).sum()
print(f"Number of records for 2015: {num_records_2015}")
print(f"Number of records for 2014: {num_records_2014}")
```

## 6.

Extract the data that pertains exclusively to 2015 and, for that subset, repeat the procedure from task 4.

```python
#TODO
#Extract the data that pertains exclusively to 2015 and, for that subset, repeat the procedure from task 4.
stocks_2015 = stocks[stocks['Date'].dt.year == 2015]
num_increase_days_2015 = (stocks_2015['ChangeRel'] > 0).sum()
num_identical_days_2015 = (stocks_2015['ChangeRel'] == 0).sum()
num_decrease_days_2015 = (stocks_2015['ChangeRel'] < 0).sum()
print(f"Number of days with a price increase in 2015: {num_increase_days_2015}")
print(f"Number of days with a price identical to the opening price in 2015: {num_identical_days_2015}")
print(f"Number of days with a price decrease in 2015: {num_decrease_days_2015}")
```

## 7.

For records from 2015, calculate the maximum opening daily stock price for the days when the closing stock price was lower than the opening price.

```python
#TODO
#For records from 2015, calculate the maximum opening daily stock price for the days when the closing stock price was lower than the opening price.
max_opening_price_2015 = stocks_2015.loc[stocks_2015['Close'] < stocks_2015['Open'], 'Open'].max()
print(f"Maximum opening daily stock price in 2015 when closing price was lower than opening price: {max_opening_price_2015}")
```

If you haven't already, solve the previous task by iterating through the dataset previously filtered using a logical condition.

```python
#TODO
#Solve the previous task by iterating through the dataset previously filtered using a logical condition.
max_opening_price_2015_iter = None
for _, row in stocks_2015.iterrows():
    if row['Close'] < row['Open']:
        if max_opening_price_2015_iter is None or row['Open'] > max_opening_price_2015_iter:
            max_opening_price_2015_iter = row['Open']
print(f"Maximum opening daily stock price in 2015 when closing price was lower than opening price (iterative): {max_opening_price_2015_iter}")
```
