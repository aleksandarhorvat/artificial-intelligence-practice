# Numpy and Matplotlib 

This notebook contains examples that combine NumPy for generating and manipulating data with Matplotlib for visualization.

```python
#!pip install matplotlib - in a virtual environment, we install the Matplotlib library.
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("NumPy and Matplotlib have been successfully loaded!")
```

## Line Charts

Displaying trends and functions.

```python
# Plotting a simple function
x = np.linspace(0, 10, 100) # 100 points between 0 and 10
print(x)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--')
plt.title('Plot of the function sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Multiple lines on the same plot
y2 = np.cos(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)', color='red')
plt.plot(x, y2, label='cos(x)', color='green', linewidth=2)
plt.title('Plot of sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## Scatter Plots

Displaying individual data points and their relationships.

```python
# Generating random points and scatter plot
np.random.seed(42)
num_points = 50
x_scatter = np.random.rand(num_points) * 10
y_scatter = 2 * x_scatter + 1 + np.random.randn(num_points) * 2 # Linear relationship with noise
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='purple', alpha=0.7, label='Data Points')
plt.title('Scatter plot of random points')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.grid(True)
plt.legend()
plt.show()
```

## Histograms

Displaying the distribution of data.

```python
# Histogram of normal distribution
data = np.random.randn(1000) # 1000 points from standard normal distribution

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of normal distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Bar charts

Comparison of categorical data.

```python
# Bar chart of sales by quarters
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
sales = np.array([150, 220, 180, 250])

plt.figure(figsize=(8, 6))
plt.bar(quarters, sales, color=['red', 'green', 'blue', 'orange'])
plt.title('Sales by quarters')
plt.xlabel('Quarter')
plt.ylabel('Sales (in thousands)')
plt.ylim(0, 300) # Setting Y-axis limit
#plt.grid() - adding grid to the plot
plt.show()
```

## Subplots

Displaying multiple plots in one figure for easier comparison.

```python
x_sub = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x_sub)
y_cos = np.cos(x_sub)
y_tan = np.tan(x_sub)

plt.figure(figsize=(12, 8))

# First subplot: Sinus
plt.subplot(2, 2, 1) # 2 rows, 2 columns, first plot
plt.plot(x_sub, y_sin, color='blue')
plt.title('Sinus function')
plt.grid(True)

# Second subplot: Cosine
plt.subplot(2, 2, 2)
plt.plot(x_sub, y_cos, color='green')
plt.title('Cosine function')
plt.grid(True)

# Third subplot: Tangent (with Y-axis limit due to asymptotes)
plt.subplot(2, 2, 3)
plt.plot(x_sub, y_tan, color='red')
plt.title('Tangent function')
plt.ylim(-5, 5) # Y-axis limit
plt.grid(True)

# Fourth subplot: Scatter plot of random data
np.random.seed(7) # New seed for this subplot
x_rand = np.random.rand(50) * 10
y_rand = np.random.rand(50) * 10
plt.subplot(2, 2, 4)
plt.scatter(x_rand, y_rand, color='orange', marker='o')
plt.title('Random Scatter plot')
plt.grid(True)

plt.tight_layout() # Automatic layout adjustment to avoid overlap
plt.show()
```

## Advanced examples - Visualizing data distribution with more bins and density


```python
# Displaying probability density function (PDF) of histograms
data_norm = np.random.normal(loc=0, scale=1, size=10000) # Normal distribution
data_exp = np.random.exponential(scale=1, size=10000) # Exponential distribution

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_norm, bins=50, density=True, alpha=0.7, color='teal', edgecolor='black')
plt.title('Histogram of normal distribution (density)')
plt.xlabel('Value')
plt.ylabel('Probability density')
plt.grid(True, linestyle='--', alpha=0.6)

plt.subplot(1, 2, 2)
plt.hist(data_exp, bins=50, density=True, alpha=0.7, color='magenta', edgecolor='black')
plt.title('Histogram of exponential distribution (density)')
plt.xlabel('Value')
plt.ylabel('Probability density')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

## 3D plot (Advanced, requires `mplot3d`)

For more complex visualizations in 3D space

```python
#!pip install mpl-tools
```

```python
# Simple 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(10)
x_3d = np.random.rand(100)
y_3d = np.random.rand(100)
z_3d = np.random.rand(100)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_3d, y_3d, z_3d, c='blue', marker='o', s=50)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('Example 3D Scatter plot')

plt.show()
```
